```python
df = pd.read_csv('./Dhana_LoanApplication_wip.csv', )
```

```python
def add_missing_info(gr):
    df = pd.DataFrame(gr).reset_index(drop=True)
    MonthlyCost = df.loc[df['Activity'] == 'O_Create Offer', 'MonthlyCost'].iloc[0]
    Selected = df.loc[df['Activity'] == 'O_Create Offer', 'Selected'].iloc[0] # missing
    FirstWithdrawalAmount =  df.loc[df['Activity'] == 'O_Create Offer', 'FirstWithdrawalAmount'].iloc[0]
    Accepted =  df.loc[df['Activity'] == 'O_Create Offer', 'Accepted'].iloc[0] # missing
    CreditScore =  df.loc[df['Activity'] == 'O_Create Offer', 'CreditScore'].iloc[0]
    NumberOfTerms =  df.loc[df['Activity'] == 'O_Create Offer', 'NumberOfTerms'].iloc[0]
    OfferedAmount = df.loc[df['Activity'] == 'O_Create Offer', 'OfferedAmount'].iloc[0]
    
    before_cr_offer = list(df.index)[0:df.index[df['Activity'] == 'O_Create Offer'].tolist()[0]]
    after_cr_offer = list(df.index)[df.index[df['Activity'] == 'O_Create Offer'].tolist()[0]:]
    
    df_before = df.iloc[before_cr_offer]
    df_before['MonthlyCost']=0
    df_before['Selected']='missing'
    df_before['FirstWithdrawalAmount']=0
    df_before['Accepted']='missing'
    df_before['CreditScore']=0
    df_before['NumberOfTerms']=0
    df_before['OfferedAmount']=0
    
    
    df_after = df.iloc[after_cr_offer]
    df_after['MonthlyCost']=df.loc[df['Activity'] == 'O_Create Offer', 'MonthlyCost'].iloc[0]
    df_after['Selected']=df.loc[df['Activity'] == 'O_Create Offer', 'Selected'].iloc[0]
    df_after['FirstWithdrawalAmount']=df.loc[df['Activity'] == 'O_Create Offer', 'FirstWithdrawalAmount'].iloc[0]
    df_after['Accepted']=df.loc[df['Activity'] == 'O_Create Offer', 'Accepted'].iloc[0]
    df_after['CreditScore']=df.loc[df['Activity'] == 'O_Create Offer', 'CreditScore'].iloc[0]
    df_after['NumberOfTerms']=df.loc[df['Activity'] == 'O_Create Offer', 'NumberOfTerms'].iloc[0]
    df_after['OfferedAmount']=df.loc[df['Activity'] == 'O_Create Offer', 'OfferedAmount'].iloc[0]

    df = pd.concat([df_before, df_after])
    return df
    
```

```python
res_df = df.groupby("Case ID").apply(add_missing_info)
res_df = res_df.reset_index(drop=True)
res_df
```

```python
case_id_col = "Case ID"
activity_col = "Activity"
resource_col = 'org:resource'
timestamp_col = 'time:timestamp'
data = res_df
import numpy as np

def extract_timestamp_features(group):
    
    group = group.sort_values(timestamp_col, ascending=False, kind='mergesort')
    
    tmp = group[timestamp_col] - group[timestamp_col].shift(-1)
    tmp = tmp.fillna(pd.Timedelta(seconds=0))
    group["timesincelastevent"] = tmp.apply(lambda x: float(x / np.timedelta64(1, 'm'))) # m is for minutes

    tmp = group[timestamp_col] - group[timestamp_col].iloc[-1]
    tmp = tmp.fillna(pd.Timedelta(seconds=0))
    group["timesincecasestart"] = tmp.apply(lambda x: float(x / np.timedelta64(1, 'm'))) # m is for minutes

    group = group.sort_values(timestamp_col, ascending=True, kind='mergesort')
    group["event_nr"] = range(1, len(group) + 1)
    
    return group

def get_wip(date):
    return sum((dt_first_last_timestamps["start_time"] <= date) & (dt_first_last_timestamps["end_time"] > date))

```

```python
import sys
case_id_col = "Case ID"
activity_col = "Activity"
resource_col = 'org:resource'
timestamp_col = 'time:timestamp'
data = res_df
    
# add event duration
data[timestamp_col] = pd.to_datetime(data[timestamp_col])
data["timesincemidnight"] = data[timestamp_col].dt.hour * 60 + data[timestamp_col].dt.minute
data["month"] = data[timestamp_col].dt.month
data["weekday"] = data[timestamp_col].dt.weekday
data["hour"] = data[timestamp_col].dt.hour

# add features extracted from timestamp
print("Extracting timestamp features...")
sys.stdout.flush()
data = data.groupby(case_id_col).apply(extract_timestamp_features)

# add inter-case features
print("Extracting wip ...")
sys.stdout.flush()
data = data.sort_values([timestamp_col], ascending=True, kind='mergesort')
dt_first_last_timestamps = data.groupby(case_id_col)[timestamp_col].agg([min, max])
dt_first_last_timestamps.columns = ["start_time", "end_time"]
data["wip"] = data[timestamp_col].apply(get_wip)
data
```

```python
def add_label(file):
    
    # function to cheack if a case contains A_Pending activity or not
    def check_A_Pending(gr):
        df = pd.DataFrame(gr)
        if  "A_Pending" in list(df['Activity']):
            df['label'] = "regular"
        else:
            df['label'] = "deviant"
        return df
    
    # read data and del old label
    df = file
    # add new label for each case based on A_Pending
    df = df.groupby('Case ID').apply(check_A_Pending)
    df = df.reset_index(drop=True)
    return df

# function to add noOfOffers column to the original data based on "O_Created" Activity
def add_No_Of_Offers(df):
    # get all observations with O_Created activity
    tmp_df = df[df['Activity'] == "O_Created"]  # to count offers
    # count numbe rof offers for each case
    tmp_df2 = pd.DataFrame(tmp_df.groupby(['Case ID'])['Activity'].count()).reset_index()
    tmp_df2.columns = ['Case ID', 'NumberOfOffers']
    df = pd.merge(tmp_df2, df, on='Case ID')
    return df 
```

```python
df = add_label(data)
df = add_No_Of_Offers(df)
df
```
